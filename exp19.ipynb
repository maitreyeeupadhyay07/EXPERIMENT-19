```python
# Treemap (Company DEpartment Budget)
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'Department': ['HR', 'IT', 'Sales', 'Marketing'],
    'Budget': [20000,50000,40000,30000]
})

fig = px.treemap(df, path=['Department'], values='Budget', title='maitreyee')
fig.show()
```

```python
# dendrogram
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
# Sample data (can represent students, features, etc)
data = np.array([
    [5,3],
    [10,15],
    [15,12],
    [24, 10],
    [30, 30]
])

linked = linkage(data, method='ward')

plt.figure(figsize=(6,4))
dendrogram(linked)

plt.title('maitreyee')
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.show()
```

```python
# venn diagram
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

A = {1, 2, 3, 4}
B = {3,4, 5, 6}

venn2(subsets=(A, B), set_labels=('Set A', 'Set B'))
plt.title('maitreyee')
plt.show()
```

```python
# sankey diagram (Student flow)
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node=dict(label=["Admission","First Year","Second Year", "Placed"]),
    link=dict(
        source=[0, 1, 2],
        target=[1, 2, 3],
        value=[100, 80, 60]
))
])

fig.show()
```

```python
# 5. 3D Scatter Plot (Student Performance)
import pandas as pd
import plotly.express as px

df = pd.DataFrame({
    'Study_Hours': [2,4,6,8,5],
    'Marks': [50,65,75,90,70],
    'Attendance': [60,75,80,90,70]
})

fig = px.scatter_3d(df,
                    x='Study_Hours',
                    y='Marks',
                    z='Attendance',
                    title="maitreyee")

fig.show()
```

```python
# 6. Radar Chart (Skill Assessment)
import plotly.graph_objects as go

skills = ['Python','ML','DBMS','DSA','Communication']
values = [4,3,5,4,3]

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=values,
    theta=skills,
    fill='toself',
    name='maitreyee'
))

fig.show()
```
